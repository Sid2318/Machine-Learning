##cleaning the data

```python
backup = car.copy()
```

```python
car = car[car['year'].str.isnumeric()]
```

```python
car['year'] = car['year'].astype(int)
```

```python
car = car[car['Price'] != 'Ask For Price']
```

```python
car['Price'] = car['Price'].str.replace(',','').astype(int)
```

```python
car['kms_driven'] = car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
car = car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven'] = car['kms_driven'].astype(int)
```

```python
car = car[~car['fuel_type'].isna()]
```

```python
car['name']
```

```python
car['name'] = car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
car.reset_index(drop=True)
```

```python
car.describe()
```

```python
car = car[car['Price'] <6e6].reset_index(drop=True)
```

```python
car.to_csv('Cleaned_car.csv')
```

##Model

```python
X = car.drop(columns='Price')
y = car['Price']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

The OneHotEncoder is a preprocessing technique used in machine learning to convert categorical data into a numerical format. It is available in scikit-learn

👉 make_column_transformer helps us do this automatically by applying the right transformation to the right columns

Once the data is prepared, we need to train a machine learning model. Normally, you would first transform the data, then train the model separately. But make_pipeline does everything in one go!

👉 It’s like a conveyor belt 🎡:

First, it transforms the data (using make_column_transformer).
Then, it feeds the cleaned data into a machine learning model (like Logistic Regression).
Finally, it makes predictions based on new data.

```python
ohe = OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
#to see max r2_score
scores = []
for i in range(1000):
  X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=i)
  lr = LinearRegression()
  pipe = make_pipeline(column_trans,lr)
  pipe.fit(X_train,y_train)
  y_pred = pipe.predict(X_test)
  scores.append(r2_score(y_test,y_pred))
```

```python
import numpy as np
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
r2_score(y_test,y_pred)
```
